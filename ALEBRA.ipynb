```python
import sympy as sp
# A matriz de ejemplo
A = sp.Matrix([[3,2,-4],[10,5,-8],[4,7,1]]) #Matriz de coeficientes
b = sp.Matrix([18,23,41]) #Matriz de términos independientes
rango=A.rank() #Determino el rango de A
U=A #Creo la matriz U
#Creo una matriz identidad para L
L = sp.eye(rango) #Crea la matriz identidad
for i in (range(rango)):
 for j in (range(i+1,rango)):
    escalar=-U[j,i]/U[i,i]
    U=U.elementary_row_op(op='n->n+km', row=j, k=escalar, row1=i)
    L=L.elementary_col_op(op='n->n+km', col=i, k=-escalar, col1=j)
sp.pprint((L,U))#pretty printer para que la salida sea más visual

#Sustitucion hacia adelante
y=b
for i in range(len(b)):
  for j in range(i):
    y[i]=y[i]-(L[i,j]*y[j])
  y[i]=y[i]/L[i,i]
  
x=y
#Susitucion hacia atrás
for i in range(len(b)-1,-1,-1):
  for j in range(len(b)-1,i,-1):
    x[i]=x[i]-(U[i,j]*x[j])
  x[i]=x[i]/U[i,i]

x #Mostramos el resultado
```

```python
import cmath
import math

def sacarRaices(c, n):
    """
    sacar_raices(complejo, indice) -> [raiz_1, raiz_2, ... , raiz_n]
    Devuelve todas las raíces de un radicando complejo con indice n
    Argumentos:
    c: radicando complejo
    n: indice de la raíz
    """
    
    plr = cmath.polar(c)
    
    angulo = plr[1]
    print (angulo)
    if angulo < 0:
        angulo += math.pi * 2

    raices = []
    raiz_radio = plr[0] ** (1. / n)
    print (raiz_radio)

    for i in range(n):
        i_theta = (angulo + 2 * math.pi * i) / n
        raices.append( cmath.rect(raiz_radio, i_theta) )

    return raices
print (sacarRaices(8+0j,3))

```

```python
import numpy as np
# INGRESO
A = np.array([[5,-2,-4],
              [3,-2,-2],
              [6,-2,-5]], dtype=float)

# PROCEDIMIENTO
casicero = 1e-15 # Considerar como 0

# matriz identidad
tamano = np.shape(A)
n = tamano[0]
identidad = np.identity(n)

# Matriz aumentada

AB = np.concatenate((A,identidad),axis=1)
AB0 = np.copy(AB)

# Pivoteo parcial por filas
tamano = np.shape(AB)
n = tamano[0]
m = tamano[1]

# Para cada fila en AB
for i in range(0,n-1,1):
    # columna desde diagonal i en adelante
    columna = abs(AB[i:,i])
    dondemax = np.argmax(columna)
    
    # dondemax no está en diagonal
    if (dondemax !=0):
        # intercambia filas
        temporal = np.copy(AB[i,:])
        AB[i,:] = AB[dondemax+i,:]
        AB[dondemax+i,:] = temporal
AB1 = np.copy(AB)

# eliminacion hacia adelante
for i in range(0,n-1,1):
    pivote = AB[i,i]
    adelante = i+1
    for k in range(adelante,n,1):
        factor = AB[k,i]/pivote
        AB[k,:] = AB[k,:] - AB[i,:]*factor
AB2 = np.copy(AB)

# elimina hacia atras
ultfila = n-1
ultcolumna = m-1
for i in range(ultfila,0-1,-1):
    pivote = AB[i,i]
    atras = i-1 
    for k in range(atras,0-1,-1):
        factor = AB[k,i]/pivote
        AB[k,:] = AB[k,:] - AB[i,:]*factor
    # diagonal a unos
    AB[i,:] = AB[i,:]/AB[i,i]

inversa = np.copy(AB[:,n:])

# SALIDA

print('Matriz aumentada:')
print(AB0)
print('Pivoteo parcial por filas')
print(AB1)
print('eliminacion hacia adelante')
print(AB2)
print('eliminación hacia atrás')
print(AB)
print('Inversa de A: ')
print(inversa)
```

```python
import cmath
import math
def sacarRaices(c, n):
    """
    sacar_raices(complejo, indice) -> [raiz_1, raiz_2, ... , raiz_n]
    Devuelve todas las raíces de un radicando complejo con indice n
    Argumentos:
    c: radicando complejo
    n: indice de la raíz
    """
    
    plr = cmath.polar(c)
    
    angulo = plr[1]
    if angulo < 0:
        angulo += math.pi * 2

    raices = []
    raiz_radio = plr[0] ** (1. / n)

    for i in range(n):
        i_theta = (angulo + 2 * math.pi * i) / n
        raices.append( cmath.rect(raiz_radio, i_theta) )

    return raices

sacarRaices(1/3+0j,4)

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
vectors=np.array([[0,0,0,4,0,3]])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for vector in vectors: 
  v = np.array([vector[3],vector[4],vector[5]]) 
  vlength=np.linalg.norm(v) 
  ax.quiver(vector[0],vector[1],vector[2],vector[3],vector[4],vector[5], pivot='tail',length=vlength,arrow_length_ratio=0.3/vlength)
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
